# Polynomial Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values  #column 1 and 2 are redundant so grab the already encoded column
y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression 
lin_rg = LinearRegression() 
lin_rg.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```python
#we are only training the model because it's a small sample and need all opf it to train the model so not test
from sklearn.preprocessing import PolynomialFeatures # poly_rg is created as an instance (object) 
#of PolynomialFeatures class
poly_rg = PolynomialFeatures(degree=4) #building a polynomial regression model entails going thru x**2, X**3, up to X**N. 
#(degree=2 means X**2 which is n=2) 
X_poly = poly_rg.fit_transform(X)
lin_rg_2 = LinearRegression()
lin_rg_2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_rg.predict(X), color ='blue') #not lin_rg_2 (that's the polynomial regression model and 
#this is asking for linear regression results)
plt.title('Truth of Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_rg_2.predict(X_poly), color ='blue') #we are working on N=2 (the squared position levels) so we 
#can't use X (N=1). X_poly, is poly_rg.fit_transform(X), and poly_rg = PolynomialFeatures(degree=2) or N=2  
#notes in chpt 88 at 55 secs in
plt.title('Truth of Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()

#I changed degree to 4 (N=4) and retrained the model to get our regression line to hit more of the points
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1) #instead of going up by 1 to the next salary we are going by .1 
#which gives us a smoother line
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_rg_2.predict(poly_rg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
